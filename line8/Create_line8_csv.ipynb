```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('C:/Users/kimmi/ExJupyterNB/data_collection.csv', encoding='utf-8')
```

```python
df
```

```python
df.columns = ['날짜', '호선', '역번호', '역명', '구분', 'a', 'b', 'c', 
              'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm',
              'n', 'o', 'p', 'q', 'r', 's']
df
```

```python
df2 = df.copy()
df2['a'] = df2.a - df2.a.shift(-1)
df2['b'] = df2.b - df2.b.shift(-1)
df2['c'] = df2.c - df2.c.shift(-1)
df2['d'] = df2.d - df2.d.shift(-1)
df2['e'] = df2.e - df2.e.shift(-1)
df2['f'] = df2.f - df2.f.shift(-1)
df2['g'] = df2.g - df2.g.shift(-1)
df2['h'] = df2.h - df2.h.shift(-1)
df2['i'] = df2.i - df2.i.shift(-1)
df2['j'] = df2.j - df2.j.shift(-1)
df2['k'] = df2.k - df2.k.shift(-1)
df2['l'] = df2.l - df2.l.shift(-1)
df2['m'] = df2.m - df2.m.shift(-1)
df2['n'] = df2.n - df2.n.shift(-1)
df2['o'] = df2.o - df2.o.shift(-1)
df2['p'] = df2.p - df2.p.shift(-1)
df2['q'] = df2.q - df2.q.shift(-1)
df2['r'] = df2.r - df2.r.shift(-1)
df2['s'] = df2.s - df2.s.shift(-1)
df2
```

```python
df2 = df2[::2]
df2
```

```python
df2 = df2.drop('구분', axis=1)
df2
```

```python
df2.to_csv("semifinal_data.csv")
```
